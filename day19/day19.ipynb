```python
with open("day19.txt", "r") as f:
    lists = [x for x in f.read().strip().split("\n\n")]

workflow, parts = lists
```

```python
import re

# Function to extract integers from a string using regular expression
def ints(s):
    return list(map(int, re.findall(r"-?\d+", s)))

# Split the 'parts' and 'workflow' strings into lists and dictionaries, respectively
parts = [ints(x) for x in parts.split("\n")]
workflow = {l.split("{")[0]: l.split("{")[1][:-1] for l in workflow.split("\n")}

# Function to recursively evaluate a workflow based on conditions
def evaluate_part(part, work):
    w = workflow[work]
    x, m, a, s = part

    for it in w.split(","):
        if it == "R":
            return False
        if it == "A":
            return True
        if ":" not in it:
            return evaluate_part(part, it)
        cond = it.split(":")[0]

        if eval(cond):
            if it.split(":")[1] == "R":
                return False
            if it.split(":")[1] == "A":
                return True
            
            return evaluate_part(part, it.split(":")[1])

# Initialize a variable to store the result of part evaluations
p1 = 0

# Loop through each part and accumulate the sum if the part evaluation is True
for part in parts:
    if evaluate_part(part, "in"):
        p1 += sum(part)

# Print the result of part evaluations
print(p1)

# Function to update acceptance ranges based on conditions
def both(ch, gt, val, ranges):
    ch = 'xmas'.index(ch)
    ranges2 = []
    for rng in ranges:
        rng = list(rng)
        lo, hi = rng[ch]
        if gt:
            lo = max(lo, val + 1)
        else:
            hi = min(hi, val - 1)
        if lo > hi:
            continue
        rng[ch] = (lo, hi)
        ranges2.append(tuple(rng))
    return ranges2

# Function to calculate acceptance ranges for the outer workflow
def acceptance_ranges_outer(work):
    return acceptance_ranges_inner(workflow[work].split(","))

# Function to calculate acceptance ranges based on inner workflow conditions
def acceptance_ranges_inner(w):
    it = w[0]
    if it == "R":
        return []
    if it == "A":
        return [((1, 4000), (1, 4000), (1, 4000), (1, 4000))]
    if ":" not in it:
        return acceptance_ranges_outer(it)
    cond = it.split(":")[0]
    gt = ">" in cond
    ch = cond[0]
    val = int(cond[2:])
    val_inverted = val + 1 if gt else val - 1
    if_cond_is_true = both(ch, gt, val, acceptance_ranges_inner([it.split(":")[1]]))
    if_cond_is_false = both(ch, not gt, val_inverted, acceptance_ranges_inner(w[1:]))
    return if_cond_is_true + if_cond_is_false

# Initialize a variable to store the result of acceptance range calculations
p2 = 0

# Loop through acceptance ranges and calculate the product of ranges
for rng in acceptance_ranges_outer('in'):
    v = 1
    for lo, hi in rng:
        v *= hi - lo + 1
    p2 += v

# Print the result of acceptance range calculations
print(p2)
```
